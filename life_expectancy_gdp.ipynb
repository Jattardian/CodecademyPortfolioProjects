```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('all_data.csv')
```

```python
print(df.head())
```

```python
df = df.apply(lambda x: x.str.replace( ' ', '') if x.dtype == 'object' else x)
```

```python
print(df.head())
```

```python
df = df.rename({"Life expectancy at birth (years)":"LEABY"}, axis = "columns")
print(df.head(16))
print(df.info())
```

```python
print(df.Country.unique())
```

```python
#Life expectancy in Chile between 2000 - 2015

plt.figure(figsize=(10, 7))
df_chile = df[df['Country'] == 'Chile']
sns.lineplot(x = range(len(df_chile)) , y='LEABY', data = df_chile,  label = 'Chile')
ax = plt.subplot()
ax.set_xticks(range(len(df_chile)))
ax.set_xticklabels(df_chile['Year'], rotation = 45)
plt.title('Life expectancy between 2000 - 2015')


#Life expectancy in China between 2000 - 2015
df_China = df[df['Country'] == 'China']
sns.lineplot(x = range(len(df_China)) , y='LEABY', data = df_China,   label = 'China')
ax = plt.subplot()
ax.set_xticks(range(len(df_China)))
ax.set_xticklabels(df_China['Year'], rotation = 45)



#Life expectancy in United States of America between 2000 - 2015
df_UnitedStatesofAmerica = df[df['Country'] == 'UnitedStatesofAmerica']
sns.lineplot(x = range(len(df_UnitedStatesofAmerica)) , y='LEABY', data = df_UnitedStatesofAmerica,  label = 'USA')
ax = plt.subplot()
ax.set_xticks(range(len(df_UnitedStatesofAmerica)))
ax.set_xticklabels(df_UnitedStatesofAmerica['Year'], rotation = 45)

#Life expectancy in Germany between 2000 - 2015
df_Germany = df[df['Country'] == 'Germany']
sns.lineplot(x = range(len(df_Germany)) , y='LEABY', data = df_Germany, label = 'Germany')
ax = plt.subplot()
ax.set_xticks(range(len(df_Germany)))
ax.set_xticklabels(df_Germany['Year'], rotation = 45)

#Life expectancy in Mexico between 2000 - 2015
df_Mexico = df[df['Country'] == 'Mexico']
sns.lineplot(x = range(len(df_Mexico)) , y='LEABY', data = df_Mexico,   label = "Mexico")
ax = plt.subplot()
ax.set_xticks(range(len(df_Mexico)))
ax.set_xticklabels(df_Mexico['Year'], rotation = 45)
plt.legend()
plt.show()
plt.clf()

#Life expectancy in Zimbabwe between 2000 - 2015
plt.figure(figsize=(10, 7))
df_Zimbabwe = df[df['Country'] == 'Zimbabwe']
sns.lineplot(x = range(len(df_Zimbabwe)) , y='LEABY', data = df_Zimbabwe)
ax = plt.subplot()
ax.set_xticks(range(len(df_Zimbabwe)))
ax.set_xticklabels(df_Zimbabwe['Year'], rotation = 45)
plt.title('Life expectancy in Zimbabwe between 2000 - 2015')

plt.show()
plt.clf()
```

```python
#GDP in Chile between 2000 - 2015

plt.figure(figsize=(10, 7))
df_chile = df[df['Country'] == 'Chile']
sns.lineplot(x = range(len(df_chile)) , y='GDP', data = df_chile,  label = 'Chile')
ax = plt.subplot()
ax.set_xticks(range(len(df_chile)))
ax.set_xticklabels(df_chile['Year'], rotation = 45)
plt.title('GDP between 2000 - 2015')


#GDP in China between 2000 - 2015
df_China = df[df['Country'] == 'China']
sns.lineplot(x = range(len(df_China)) , y='GDP', data = df_China,   label = 'China')
ax = plt.subplot()
ax.set_xticks(range(len(df_China)))
ax.set_xticklabels(df_China['Year'], rotation = 45)



#GDP in United States of America between 2000 - 2015
df_UnitedStatesofAmerica = df[df['Country'] == 'UnitedStatesofAmerica']
sns.lineplot(x = range(len(df_UnitedStatesofAmerica)) , y='GDP', data = df_UnitedStatesofAmerica,  label = 'USA')
ax = plt.subplot()
ax.set_xticks(range(len(df_UnitedStatesofAmerica)))
ax.set_xticklabels(df_UnitedStatesofAmerica['Year'], rotation = 45)

#GDP in Germany between 2000 - 2015
df_Germany = df[df['Country'] == 'Germany']
sns.lineplot(x = range(len(df_Germany)) , y='GDP', data = df_Germany, label = 'Germany')
ax = plt.subplot()
ax.set_xticks(range(len(df_Germany)))
ax.set_xticklabels(df_Germany['Year'], rotation = 45)

#GDP in Mexico between 2000 - 2015
df_Mexico = df[df['Country'] == 'Mexico']
sns.lineplot(x = range(len(df_Mexico)) , y='GDP', data = df_Mexico,   label = "Mexico")
ax = plt.subplot()
ax.set_xticks(range(len(df_Mexico)))
ax.set_xticklabels(df_Mexico['Year'], rotation = 45)
plt.legend()


#GDP in Zimbabwe between 2000 - 2015

df_Zimbabwe = df[df['Country'] == 'Zimbabwe']
sns.lineplot(x = range(len(df_Zimbabwe)) , y='GDP', data = df_Zimbabwe, label = 'Zimbabwe')
ax = plt.subplot()
ax.set_xticks(range(len(df_Zimbabwe)))
ax.set_xticklabels(df_Zimbabwe['Year'], rotation = 45)
plt.ylabel('GDP per capita in 1000USD')

plt.show()
plt.clf()
```

```python
#correlation between life expencanxy and GDP in a Zimbabwe between 2000 - 2015
plt.figure(figsize = (10, 7 ))

plt.scatter(df_Zimbabwe.GDP, df_Zimbabwe.LEABY)
plt.xlabel('GDP per capita in 1000USD')
plt.title("life expencanxy and GDP in a Zimbabwe between 2000 - 2015")
plt.ylabel('Year')

plt.show()
plt.clf()
```

```python
#whats the average life expectanxy in these 6 nations
dfMeans = df.groupby("Country").mean().reset_index()
print(dfMeans)
plt.bar(dfMeans.Country, dfMeans.LEABY)
ax = plt.subplot()
ax.set_xticks(range(6))
ax.set_xticklabels(dfMeans['Country'], rotation = 45 )
plt.axis([-1 , 6, 40, 90])
plt.title('Average Life expectancy per nation')
plt.ylabel("Years")
plt.show()
plt.clf()
```
